```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from resnet18_32x32 import ResNet18_32x32


from torchvision import transforms
from sklearn.manifold import TSNE

# 
model = ResNet18_32x32()
model.load_state_dict(torch.load('./resnet/best.ckpt'))
output = "./output_prob"


# CIFAR-10
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=1,
                                         shuffle=False, num_workers=2)


# 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)


# CIFAR-10
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
for number in range(10):
    # 
    model.eval()
    class_probs = []
    class_labels = []
    with torch.no_grad():
        for images, labels in testloader:
            if labels.item() == number:
                images = images.to(device)
                labels = labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                probabilities = torch.softmax(outputs, dim=1) 
                class_probs.append(probabilities.cpu().numpy()[0]) #
                class_labels.append(predicted.item())
    
    class_probs = np.array(class_probs)
    class_labels = np.array(class_labels)
    
    latent_dim = 100
    num_classes = 10
    lr = 0.0002
    batch_size = 64
    num_epochs = 20
    
    
    class Generator(nn.Module):
        def __init__(self, latent_dim, num_classes):
            super(Generator, self).__init__()
            self.label_embed = nn.Embedding(num_classes, latent_dim)
            self.generator = nn.Sequential(
                nn.Linear(latent_dim * 2, 128),  # 
                nn.LeakyReLU(0.2, inplace=True),
                nn.Linear(128, 256),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Linear(256, 512),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Linear(512, num_classes),
                nn.Softmax(dim=1)
            )
    
        def forward(self, labels, noise):
            gen_input = self.label_embed(labels)
            gen_input_with_noise = torch.cat((gen_input, noise), dim=1)  # 
            class_probs = self.generator(gen_input_with_noise)
            return class_probs
    
    # 
    generator = Generator(latent_dim, num_classes).to(device)
    generator.load_state_dict(torch.load('./output/best/32/generator100+100.pth'))  # Xepoch
    generated_probs = []
    # 
    generator.eval()
    with torch.no_grad():
        for i in range(1000):
            gen_labels = torch.tensor([number]).to(device)
            noise = torch.randn(1, latent_dim).to(device)
            gen_class_probs = generator(gen_labels, noise)
            
            generated_probs.append(gen_class_probs.cpu().numpy()[0])
    generated_probs = np.array(generated_probs)
#    for i in range(100):
#        gen_probs = generated_probs[i]
#        print(f"Element {i+1}:")
#        print("gen_Probabilities:", gen_probs)
#        print()
    
        
    
    all_probs = np.concatenate((class_probs, generated_probs))
    all_labels = np.concatenate((np.zeros(len(class_probs)), np.ones(len(generated_probs))))
    
        
        
    # T-SNE

    tsne = TSNE(n_components=2, random_state=42)
    tsne_result = tsne.fit_transform(all_probs)
    
    # 
    colors = ['blue', 'red']
    all_labels = all_labels.astype(int)
    # 
    edgecolors = [colors[label] for label in all_labels]

    # 
    plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c='none', marker='o', edgecolors=edgecolors)
    plt.title('T-SNE Visualization of Class Probability Distributions (Label {})'.format(number))
    plt.show()
```
